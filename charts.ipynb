```python
mappings = {
    'v1': 'Static Partitioning',
    'v2': 'Dynamic Partitioning',
    'v3': 'In-block Compression',
    'v4': 'Cache-conscious Compression',
    'v5': 'CPU Optimizations',
    'uncompressed': 'Uncompressed',
    'simd-ized': 'Uncompresed w/ SIMD'
}
```

```python
labels = []
for v in ratios:
    if v != 'v4' and v != 'v5':
        x, y = zip(*sorted(ratios[v].items()))
        plt.plot(x, y, linewidth=4.0)
        if v in mappings:
            v = mappings[v]
        labels.append(v)
plt.legend(labels, loc='center right')
plt.title('Compression Ratio by Version')
plt.xlabel('Element Count', fontsize=22)
plt.ylabel('Compression Ratio', fontsize=22)
plt.xscale('log')
plt.show()
```

```python
labels = []
for v in comp_speeds:
    x, y = zip(*sorted(comp_speeds[v].items()))
    plt.plot(x, y, linewidth=4.0)
    if v in mappings:
        v = mappings[v]
    labels.append(v)
plt.legend(labels, loc='upper right')
plt.title('Compression Performance')
plt.xlabel('Element Count', fontsize=22)
plt.ylabel('Element Compression Latency (ns)', fontsize=22)
plt.xscale('log')
plt.show()
```
